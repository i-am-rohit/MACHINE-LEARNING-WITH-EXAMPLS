K-means clustering is a popular unsupervised machine learning algorithm for grouping data points into a predefined number of clusters.<br><br>

Core Idea:<br>

K Selection: You specify the desired number of clusters (K) to partition the data into.<br>
Initial Centroids: The algorithm randomly selects K data points as initial cluster centroids (means). These represent the tentative center of each cluster.<br>
Assignment: Each data point is assigned to the cluster with the nearest centroid based on a distance metric (often Euclidean distance).<br>
Centroid Update: The centroids are recomputed as the mean of the data points assigned to each cluster.<br>
Iteration: Steps 3 and 4 are repeated until a stopping criterion is met. This criterion can be:<br>
No significant change in cluster assignments between iterations (convergence).
A maximum number of iterations is reached.

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Sample data (replace with your actual data)
X = [[1, 2], [3, 4], [1.5, 1.8], [5, 8], [8, 8], [1, 0.6], [9, 11]]

# Define the number of clusters (K)
K = 3

# Create a KMeans model and perform clustering
kmeans = KMeans(n_clusters=K, random_state=42)
kmeans.fit(X)

# Get cluster labels for each data point
cluster_labels = kmeans.labels_

# Get the centroids of each cluster
centroids = kmeans.cluster_centers_

# Visualization setup
colors = ['red', 'green', 'blue']  # Define colors for clusters

# Plot data points with colors based on cluster labels
plt.scatter([x[0] for x in X], [x[1] for x in X], c=[colors[l] for l in cluster_labels])

# Plot centroids as larger markers with different shapes
plt.scatter(centroids[:, 0], centroids[:, 1], marker='o', s=100, linewidths=2, c='black')

# Add labels and title
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title(f'K-means Clustering (K={K})')
plt.legend()
plt.show()
```
